```python
import matplotlib.pyplot as plt
# Checking for outliers in numerical columns

columns = boston.columns

for i in columns:
    if boston[f'{i}'].dtype == 'int64':
        sns.boxplot(x=i, data=boston)
        plt.show()
 
```

```python
#  Eliminating the probable ouliers




def outlier(columns):
  column = boston[columns]
  column = pd.DataFrame()
  column[f'{columns}'] = boston[f'{columns}']
  column['upperlimit'] = boston[f'{columns}'].mean() + (boston[f'{columns}'].std() *3)
  column['lowerlimit'] = boston[f'{columns}'].mean()- (boston[f'{columns}'].std()*3)
  plt.figure(figsize=(20,7))
  column.plot()
  plt.legend()
  boston.drop(boston.loc[boston[f'{columns}'] >= column['upperlimit'].mean()].index, inplace = True)
  # boston[f'{columns}'].plot()
  



```

```python
for col in columns:
  if boston[f'{col}'].dtype == 'int64':
    outlier(col)
```
